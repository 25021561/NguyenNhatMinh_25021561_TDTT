#### Bài 2

```python
# Bài 2
import math

pi = 3.14

a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

dt_hcn = a * b

r = a / 2

dt_tron = pi * (r ** 2)

dt_cay = dt_hcn - dt_tron

print(f"Diện tích phần trồng cây còn lại là: {dt_cay:.2f}")
```

#### Bài 6

```python
# Bài 6
import math

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    p = (a + b + c) / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Diện tích tam giác là: {dien_tich:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

#### Bài 8

```python
# Bài 8: Tính tiền điện
ho_ten = input("Ten chu ho: ")
cs_truoc = int(input("Chi so thang truoc: "))
cs_nay = int(input("Chi so thang nay: "))

so_kwh = cs_nay - cs_truoc
tong_tien = 0

print(f"--- Hoá đơn cho: {ho_ten} ---")
print(f"Tổng số kWh tiêu thụ: {so_kwh}")

# Tính tiền theo phương pháp "trừ dần" cho dễ
so_kwh_tinh_tien = so_kwh

# Bậc 6: Từ 401 trở đi
if so_kwh_tinh_tien > 400:
    dien_bac_6 = so_kwh_tinh_tien - 400
    tien_bac_6 = dien_bac_6 * 3460
    tong_tien += tien_bac_6
    so_kwh_tinh_tien = 400 # Đặt lại số kWh cho bậc dưới
    print(f"Bậc 6: {dien_bac_6} kWh * 3,460 = {tien_bac_6} VND")

# Bậc 5: 301 - 400 (100 kWh)
if so_kwh_tinh_tien > 300:
    dien_bac_5 = so_kwh_tinh_tien - 300
    tien_bac_5 = dien_bac_5 * 3350
    tong_tien += tien_bac_5
    so_kwh_tinh_tien = 300
    print(f"Bậc 5: {dien_bac_5} kWh * 3,350 = {tien_bac_5} VND")

# Bậc 4: 201 - 300 (100 kWh)
if so_kwh_tinh_tien > 200:
    dien_bac_4 = so_kwh_tinh_tien - 200
    tien_bac_4 = dien_bac_4 * 2998
    tong_tien += tien_bac_4
    so_kwh_tinh_tien = 200
    print(f"Bậc 4: {dien_bac_4} kWh * 2,998 = {tien_bac_4} VND")

# Bậc 3: 101 - 200 (100 kWh)
if so_kwh_tinh_tien > 100:
    dien_bac_3 = so_kwh_tinh_tien - 100
    tien_bac_3 = dien_bac_3 * 2380
    tong_tien += tien_bac_3
    so_kwh_tinh_tien = 100
    print(f"Bậc 3: {dien_bac_3} kWh * 2,380 = {tien_bac_3} VND")

# Bậc 2: 51 - 100 (50 kWh)
if so_kwh_tinh_tien > 50:
    dien_bac_2 = so_kwh_tinh_tien - 50
    tien_bac_2 = dien_bac_2 * 2050
    tong_tien += tien_bac_2
    so_kwh_tinh_tien = 50
    print(f"Bậc 2: {dien_bac_2} kWh * 2,050 = {tien_bac_2} VND")

# Bậc 1: 0 - 50 (50 kWh)
# (so_kwh_tinh_tien lúc này luôn <= 50)
tien_bac_1 = so_kwh_tinh_tien * 1984
tong_tien += tien_bac_1
print(f"Bậc 1: {so_kwh_tinh_tien} kWh * 1,984 = {tien_bac_1} VND")

print(f"Tổng tiền điện (chưa VAT): {tong_tien} VND")

# Tính VAT 8%
tien_vat = tong_tien * 0.08
print(f"Tiền VAT (8%): {tien_vat:.1f} VND")

# Tổng tiền phải trả, làm tròn đến đơn vị đồng
thanh_toan = round(tong_tien + tien_vat)

# In kết quả cuối cùng theo mẫu
print("---")
print(f"Ho va ten: {ho_ten}")
print(f"Tien phai tra la: {thanh_toan}")
```
